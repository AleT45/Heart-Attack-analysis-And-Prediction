```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Set the figure size for plots

plt.rcParams['figure.figsize'] = (16, 8)
```

```python
#Set the style for plots

plt.style.use('fivethirtyeight')
```

```python
# Load the heart dataset

df_heart = pd.read_csv('heart.csv')
```

```python
# Print the first five rows 

print(df_heart.head())
```

```python
#Print the dimensions 

print(df_heart.shape)
```

```python
#Print the information 

print(df_heart.info())
```

```python
#Check for missing values 

print(df_heart.isnull().sum().sort_values(ascending=False))
```

```python
#Generate summary statistics 

print(df_heart.describe())
```

```python
#Check for duplicated rows

print(df_heart.duplicated().sum())
```

```python
#Print the number of unique values

print(df_heart.nunique())
```

```python
#Generate a heatmap between the columns 

plt.figure(figsize = (16,16))
sns.heatmap(df_heart.corr(), center=0, square=True, annot=True, lw=0.3)
plt.title('Correlations between Columns', fontsize=20)
plt.show()
```

```python
# The relationship range age

sns.countplot(x='age', data=df_heart)
plt.title('Range of People by Age', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('People', fontsize=15)
plt.show()
```

```python
#The relationship between age and the target colum

sns.countplot(x=df_heart['age'], hue=df_heart['output'])
plt.title('Heart Attacks by Age', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Heart Attacks', fontsize=15)
plt.show()
```

```python
#The relationship between gender

sex_data = df_heart.sex.map({1: 'Male', 0: 'Female'})
target_data = df_heart.output.map({1: 'Higher', 
                                   0: 'Lower'})
sns.countplot(hue=target_data, x=sex_data, data=df_heart)
plt.title('Chance of Heart Attacks by Gender', fontsize=20)
plt.xlabel('Gender', fontsize=15)
plt.ylabel('Heart Attacks', fontsize=15)
plt.show()
```

```python
#Blood Sugar effects

target_data = df_heart.output.map({1: 'Higher chances of heart attack', 0: 'Lesser chances of heart attack'})
sns.barplot(x = 'fbs',y = target_data, data = df_heart)
plt.title("Distribution of Blood Sugar with to Heart Attack" , fontsize= 20)
plt.xlabel('fbs',fontsize = 15)
plt.ylabel('output',fontsize = 15)
plt.xticks();df_heart
```

```python
#Chest Pain

orders = (df_heart.groupby(['cp']).mean().sort_values(by ='output', ascending = False)).index
sns.barplot(x = 'cp', y = 'output', data = df_heart, order = orders)
plt.title("Chest Pain" , fontsize= 20)
plt.xlabel('cp',fontsize = 15)
plt.ylabel('output',fontsize = 15);
```

```python
#Marjor Blood Verssels

target_data = df_heart.output.map({1: 'Higher', 0: 'Lower'})
sns.countplot(x = 'caa', hue = target_data, data = df_heart)
plt.title("Major Blood Vessels" , fontsize= 20)
plt.xlabel('caa',fontsize = 15)
plt.ylabel('output',fontsize = 15);
```

```python
#Resting blood pressure (in mm Hg)

cp_data = df_heart.cp.map({0:'Typical Angina',1:'Atypical Angina',2:'Non-Anginal Pain',3:'Asymptomatic'})
sns.violinplot(x = cp_data, y= 'trtbps', data = df_heart)
plt.title('Resting Blood Pressure (mm Hg) vs Chest Pain', fontsize = 20)
plt.xlabel('Chest Pain',fontsize = 15)
plt.ylabel('Resting Blood Pressure',fontsize = 15)
plt.show();
```

```python
# Maximum heart rate 

sns.histplot(x = 'thalachh', data = df_heart)
plt.title(' Max Heart Bits Count When Having a Heart Attack', fontsize = 20)
plt.xlabel('Heart Rate', fontsize = 15);
```

```python
sns.histplot(x = 'oldpeak', data = df_heart)

plt.xlabel('Previous peak')
```

```python
target_data = df_heart.output.map({0:'Less',1:'More'})
sns.countplot(x = target_data,data = df_heart)
plt.ylabel('People')
plt.xlabel('Suffured or Not')
plt.title('Gotten Geart Attack', fontsize = 20)
```

```python
#Blood pressure 

sns.distplot(x=df_heart['trtbps'],hist_kws=dict(edgecolor="k", linewidth=2))
plt.xlabel('Blood pressure')
plt.ylabel('Density of Blood Pressue')
plt.title('Blood Pressure to Intensity of Blood Pressure')
```

```python
#cholesterol rate

sns.distplot(x= df_heart['chol'],hist_kws=dict(edgecolor="k", linewidth=2))
plt.xlabel('Cholesterol Rate')
plt.ylabel('Density of cholesterol Rate')
plt.title('Cholesterol Rate to Intensity of Cholesterol Rate')
plt.show()
```
